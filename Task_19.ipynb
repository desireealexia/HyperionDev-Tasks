```python
# Use the data in the relevant columns to determine how age affects insurance costs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Import the data
df = pd.read_csv("insurance.csv")
print(df)
```

```python
# Plot a scatter plot with age on the x-axis and charges on the y-axis
x = df.iloc[:,:1].values
y = df.iloc[:,-1].values

plt.scatter(x,y,color = 'y')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
plt.close()
```

```python
# Using linerar_model.LinearRegression() from sklearn, fit a model to your data, and make preditions on data
linreg_model = LinearRegression()
linreg_model.fit(x,y)

y_pred = linreg_model.predict(x)

unknown_value_x = [[70]]

x_pred = np.append(x, unknown_value_x).reshape(-1,1)
y_pred = linreg_model.predict(x_pred)
```

```python
# Plot another scatter plot with the best-fit line
plt.scatter(x,y,color = 'y')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```
